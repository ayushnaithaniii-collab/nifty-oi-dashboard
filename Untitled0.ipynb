```python
import pandas as pd

rows = []
for item in data3['data']:
    rows.append({
        'strike': item['strike_price'],
        'call_oi': item['call_options']['market_data']['oi'],
        'put_oi': item['put_options']['market_data']['oi'],
        'call_ltp': item['call_options']['market_data']['ltp'],
        'put_ltp': item['put_options']['market_data']['ltp'],
    })

df = pd.DataFrame(rows)
print(df.head(10))
```

```python
total_call_oi = df['call_oi'].sum()
total_put_oi = df['put_oi'].sum()
pcr = total_put_oi / total_call_oi

print("Total Call OI:", total_call_oi)
print("Total Put OI:", total_put_oi)
print("PCR:", round(pcr, 2))

if pcr > 1.2:
    print("Signal: BULLISH - Put writers active")
elif pcr < 0.8:
    print("Signal: BEARISH - Call writers active")
else:
    print("Signal: NEUTRAL")
```

```python
max_pain_strike = None
min_pain = float('inf')

for index, row in df.iterrows():
    strike = row['strike']

    # Total loss to option writers at this strike
    call_pain = ((df[df['strike'] < strike]['strike'] - strike).abs() * df[df['strike'] < strike]['call_oi']).sum()
    put_pain = ((df[df['strike'] > strike]['strike'] - strike).abs() * df[df['strike'] > strike]['put_oi']).sum()

    total_pain = call_pain + put_pain

    if total_pain < min_pain:
        min_pain = total_pain
        max_pain_strike = strike

print("Max Pain Strike:", max_pain_strike)
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

token = "YOUR_TOKEN_HERE"

headers = {
    "Authorization": f"Bearer {token}",
    "Accept": "application/json"
}

# Fetch options chain
response = requests.get(
    "https://api.upstox.com/v2/option/chain",
    headers=headers,
    params={
        "instrument_key": "NSE_INDEX|Nifty 50",
        "expiry_date": "2026-06-16"
    }
)

data3 = response.json()

# Build dataframe
rows = []
for item in data3['data']:
    rows.append({
        'strike': item['strike_price'],
        'call_oi': item['call_options']['market_data']['oi'],
        'put_oi': item['put_options']['market_data']['oi'],
        'call_ltp': item['call_options']['market_data']['ltp'],
        'put_ltp': item['put_options']['market_data']['ltp'],
    })

df = pd.DataFrame(rows)

# PCR
total_call_oi = df['call_oi'].sum()
total_put_oi = df['put_oi'].sum()
pcr = round(total_put_oi / total_call_oi, 2)

# Max Pain
max_pain_strike = None
min_pain = float('inf')
for index, row in df.iterrows():
    strike = row['strike']
    call_pain = ((df[df['strike'] < strike]['strike'] - strike).abs() * df[df['strike'] < strike]['call_oi']).sum()
    put_pain = ((df[df['strike'] > strike]['strike'] - strike).abs() * df[df['strike'] > strike]['put_oi']).sum()
    total_pain = call_pain + put_pain
    if total_pain < min_pain:
        min_pain = total_pain
        max_pain_strike = strike

# Spot price
nifty = yf.Ticker("^NSEI")
spot = round(nifty.history(period="1d")['Close'].iloc[-1])

# Plot
plt.figure(figsize=(14, 6))
plt.bar(df['strike'], df['call_oi'], color='red', alpha=0.7, label='Call OI')
plt.bar(df['strike'], df['put_oi'], color='green', alpha=0.7, label='Put OI')
plt.axvline(x=spot, color='yellow', linestyle='--', label=f'Spot: {spot}')
plt.axvline(x=max_pain_strike, color='white', linestyle='--', label=f'Max Pain: {max_pain_strike}')
plt.title(f'Nifty OI | PCR: {pcr} | Max Pain: {max_pain_strike}')
plt.xlabel('Strike Price')
plt.ylabel('Open Interest')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print(f"Spot: {spot} | Max Pain: {max_pain_strike} | PCR: {pcr}")
```

```python
import time
from datetime import datetime

pcr_history = []

print("Recording PCR every 30 seconds... (will run 5 times)")

for i in range(5):
    response = requests.get(
        "https://api.upstox.com/v2/option/chain",
        headers=headers,
        params={
            "instrument_key": "NSE_INDEX|Nifty 50",
            "expiry_date": "2026-06-16"
        }
    )

    data = response.json()

    rows = []
    for item in data['data']:
        rows.append({
            'strike': item['strike_price'],
            'call_oi': item['call_options']['market_data']['oi'],
            'put_oi': item['put_options']['market_data']['oi'],
        })

    temp_df = pd.DataFrame(rows)
    current_pcr = round(temp_df['put_oi'].sum() / temp_df['call_oi'].sum(), 2)
    current_time = datetime.now().strftime("%H:%M:%S")

    pcr_history.append({'time': current_time, 'pcr': current_pcr})
    print(f"{current_time} — PCR: {current_pcr}")

    if i < 4:
        time.sleep(30)

print("Done.")
```

```python
pcr_df = pd.DataFrame(pcr_history)

plt.figure(figsize=(10, 4))
plt.plot(pcr_df['time'], pcr_df['pcr'], color='yellow', linewidth=2, marker='o')
plt.axhline(y=1.2, color='green', linestyle='--', alpha=0.5, label='Bullish (1.2)')
plt.axhline(y=0.8, color='red', linestyle='--', alpha=0.5, label='Bearish (0.8)')
plt.axhline(y=1.0, color='white', linestyle='--', alpha=0.3, label='Neutral (1.0)')
plt.title('PCR Trend')
plt.xlabel('Time')
plt.ylabel('PCR')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ============================================
# NIFTY OI DASHBOARD
# Built by: Aayush Naithanii
# Description: Live Nifty Options OI Analysis
# Data Source: Upstox API
# ============================================

import requests
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from datetime import datetime

# ---- CONFIG ----
TOKEN = "YOUR_TOKEN_HERE"
EXPIRY = "2026-06-16"
SYMBOL = "NSE_INDEX|Nifty 50"

# ---- FETCH DATA ----
def fetch_options_chain(token, symbol, expiry):
    headers = {
        "Authorization": f"Bearer {token}",
        "Accept": "application/json"
    }
    response = requests.get(
        "https://api.upstox.com/v2/option/chain",
        headers=headers,
        params={"instrument_key": symbol, "expiry_date": expiry}
    )
    return response.json()

# ---- BUILD DATAFRAME ----
def build_dataframe(data):
    rows = []
    for item in data['data']:
        rows.append({
            'strike': item['strike_price'],
            'call_oi': item['call_options']['market_data']['oi'],
            'put_oi': item['put_options']['market_data']['oi'],
            'call_ltp': item['call_options']['market_data']['ltp'],
            'put_ltp': item['put_options']['market_data']['ltp'],
        })
    return pd.DataFrame(rows)

# ---- CALCULATE PCR ----
def calculate_pcr(df):
    return round(df['put_oi'].sum() / df['call_oi'].sum(), 2)

# ---- CALCULATE MAX PAIN ----
def calculate_max_pain(df):
    max_pain_strike = None
    min_pain = float('inf')
    for _, row in df.iterrows():
        strike = row['strike']
        call_pain = ((df[df['strike'] < strike]['strike'] - strike).abs() * df[df['strike'] < strike]['call_oi']).sum()
        put_pain = ((df[df['strike'] > strike]['strike'] - strike).abs() * df[df['strike'] > strike]['put_oi']).sum()
        total_pain = call_pain + put_pain
        if total_pain < min_pain:
            min_pain = total_pain
            max_pain_strike = strike
    return max_pain_strike

# ---- GET SPOT PRICE ----
def get_spot():
    nifty = yf.Ticker("^NSEI")
    return round(nifty.history(period="1d")['Close'].iloc[-1])

# ---- PLOT DASHBOARD ----
def plot_dashboard(df, spot, max_pain, pcr):
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

    # Chart 1 - OI by Strike
    ax1.bar(df['strike'], df['call_oi'], color='red', alpha=0.7, label='Call OI')
    ax1.bar(df['strike'], df['put_oi'], color='green', alpha=0.7, label='Put OI')
    ax1.axvline(x=spot, color='yellow', linestyle='--', label=f'Spot: {spot}')
    ax1.axvline(x=max_pain, color='white', linestyle='--', label=f'Max Pain: {max_pain}')
    ax1.set_title(f'Nifty OI by Strike | PCR: {pcr} | Max Pain: {max_pain}')
    ax1.set_xlabel('Strike Price')
    ax1.set_ylabel('Open Interest')
    ax1.legend()
    ax1.tick_params(axis='x', rotation=45)

    # Chart 2 - PCR Signal
    colors = {'BULLISH': 'green', 'BEARISH': 'red', 'NEUTRAL': 'yellow'}
    if pcr > 1.2:
        signal = 'BULLISH'
    elif pcr < 0.8:
        signal = 'BEARISH'
    else:
        signal = 'NEUTRAL'

    ax2.barh(['PCR'], [pcr], color=colors[signal], alpha=0.8)
    ax2.axvline(x=1.2, color='green', linestyle='--', label='Bullish (1.2)')
    ax2.axvline(x=0.8, color='red', linestyle='--', label='Bearish (0.8)')
    ax2.axvline(x=1.0, color='white', linestyle='--', alpha=0.5, label='Neutral (1.0)')
    ax2.set_title(f'PCR Signal: {signal} ({pcr})')
    ax2.set_xlim(0, 2)
    ax2.legend()

    plt.suptitle(f'Nifty Options Dashboard | {datetime.now().strftime("%d %b %Y %H:%M")}',
                 fontsize=14, y=1.02)
    plt.tight_layout()
    plt.show()

# ---- RUN ----
print("Fetching live data...")
raw_data = fetch_options_chain(TOKEN, SYMBOL, EXPIRY)
df = build_dataframe(raw_data)
pcr = calculate_pcr(df)
max_pain = calculate_max_pain(df)
spot = get_spot()

print(f"Spot: {spot} | Max Pain: {max_pain} | PCR: {pcr}")
plot_dashboard(df, spot, max_pain, pcr)
```
